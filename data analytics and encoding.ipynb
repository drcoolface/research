```python
def df_to_X_y(df, window_size= window_size):
  df_as_np = df.to_numpy()
  X = []
  y = []
  for i in range(len(df_as_np)-window_size):
    row = [r for r in df_as_np[i:i+window_size]]
    X.append(row)
    label = [df_as_np[i+window_size][0], df_as_np[i+window_size][1],df_as_np[i+window_size][2], df_as_np[i+window_size][3]]
    y.append(label)
  return np.array(X), np.array(y)
```

```python
T2M_training_mean = np.mean(X_train[:, :, 0])
T2M_training_std = np.std(X_train[:, :, 0])

QV2M_training_mean = np.mean(X_train[:, :, 1])
QV2M_training_std = np.std(X_train[:, :, 1])

PRECTOTCORR_training_mean = np.mean(X_train[:, :, 2])
PRECTOTCORR_training_std = np.std(X_train[:, :, 2])

WS_training_mean = np.mean(X_train[:, :, 3])
WS_training_std = np.std(X_train[:, :, 3])

def preprocess(X):
    X[:, :, 0] = (X[:, :, 0] - T2M_training_mean) / T2M_training_std
    X[:, :, 1] = (X[:, :, 1] - QV2M_training_mean) / QV2M_training_std
    X[:, :, 2] = (X[:, :, 2] - PRECTOTCORR_training_mean) / PRECTOTCORR_training_std
    X[:, :, 3] = (X[:, :, 3] - WS_training_mean) / WS_training_std
    return X

def preprocess_output(y):
    y[:, 0] = (y[:, 0] - T2M_training_mean) / T2M_training_std
    y[:, 1] = (y[:, 1] - QV2M_training_mean) / QV2M_training_std
    y[:, 2] = (y[:, 2] - PRECTOTCORR_training_mean) / PRECTOTCORR_training_std
    y[:, 3] = (y[:, 3] - WS_training_mean) / WS_training_std
    return y
```

```python
def iterative_pred(model, exog , h):
    
    y_preds= []
    
    current_X = exog
    
    y_preds = pd.DataFrame(index=xyz[:h].index, columns=['T2M', 'QV2M', 'PRECTOTCORR', 'PS', 'Hour_sin', 'Hour_cos',
                                                         'Day_sin', 'Day_cos','Month_sin', 'Month_cos'])
    y_preds['Hour_sin'] = xyz['Hour_sin']
    y_preds['Hour_cos'] = xyz['Hour_cos']
    
    y_preds['Day_sin'] = xyz['Day_sin']
    y_preds['Day_cos'] = xyz['Day_cos']
    
    y_preds['Month_sin'] = xyz['Month_sin']
    y_preds['Month_cos'] = xyz['Month_cos']
    
    
    
    for i in range(h):
        y_pred = model.predict(current_X)
        
        y_preds.iloc[[i],[0]] = y_pred[0,0]
        y_preds.iloc[[i],[1]] = y_pred[0,1]
        y_preds.iloc[[i],[2]] = y_pred[0,2]
        y_preds.iloc[[i],[3]] = y_pred[0,3]
        
        currrent_X = np.roll(current_X, shift=-1)
        current_X[-1]= y_preds.iloc[[i]]
        
    return np.array(y_preds.iloc[:,0:4])
    
    
```

```python
def return_predictions(predictions, labels):
    y = labels
    T2M_preds, QV2M_preds, PRECTOTCORR_preds, PS_preds  = postprocess_T2M(predictions[:, 0]), postprocess_QV2M(predictions[:, 1]),  postprocess_PRECTOTCORR(predictions[:, 2]),  postprocess_PS(predictions[:, 3])
    T2M_actuals, QV2M_actuals, PRECTOTCORR_actuals, PS_actuals = postprocess_T2M(y[:, 0]), postprocess_QV2M(y[:, 1]) , postprocess_PRECTOTCORR(y[:, 2]), postprocess_PS(y[:, 3])
    df = pd.DataFrame(data={'Temperature Predictions': T2M_preds,
                        'Temperature Actuals':T2M_actuals,
                            
                          'Humidity Predictions': QV2M_preds,
                          'Humidity Actuals': QV2M_actuals,
                            
                            
                             'Precipitation Predictions': PRECTOTCORR_preds,
                          'Precipitation Actuals': PRECTOTCORR_actuals,
                            
                             'Pressure Predictions': PS_preds,
                          'Pressure Actuals': PS_actuals
                          })
    plt.figure(figsize=(200,200))
    plt.subplot(4,1,1)
    plt.plot(df['Temperature Predictions'])
    plt.plot(df['Temperature Actuals'])
    
    
    plt.figure(figsize=(200,200))
    plt.subplot(4,1,2)
    plt.plot(df['Humidity Predictions'])
    plt.plot(df['Humidity Actuals'])
    
    
    plt.figure(figsize=(200,200))
    plt.subplot(4,1,3)
    plt.plot(df['Precipitation Predictions'])
    plt.plot(df['Precipitation Actuals'])
    
    plt.figure(figsize=(200,200))
    plt.subplot(4,1,4)
    plt.plot(df['Pressure Predictions'])
    plt.plot(df['Pressure Actuals'])
    
    return df
```
